## **Extracting and merging wanted columns**

```python
# Preprocessing and selecting relevant columns
def preprocess_dataset(dataset, text_column):
    # Check if text_column is a list of columns to concatenate
    if isinstance(text_column, list):
        # Concatenate the specified columns into a single text column, converting None to empty strings
        concatenated_texts = dataset['train'].map(lambda example: {'text': ' '.join([str(example[col]) if example[col] is not None else '' for col in text_column])})
        # Create a DataFrame with the concatenated text
        df = pd.DataFrame(concatenated_texts['text'], columns=['text'])
    else:
        # If text_column is a single column, just select it
        df = pd.DataFrame(dataset['train'][text_column], columns=['text'])

    return df

# Preprocess each dataset
df1 = preprocess_dataset(dataset1, 'Merged')
df2 = preprocess_dataset(dataset2, 'text')
df3 = preprocess_dataset(dataset3, ['Context','Response'])
df4 = preprocess_dataset(dataset4, 'text')
df5 = preprocess_dataset(dataset5, 'text')
df6 = preprocess_dataset(dataset6, 'body')
df7 = preprocess_dataset(dataset7, ['Questions', 'Answers'])
df8 = preprocess_dataset(dataset8, ['Context', 'Knowledge', 'Response'])
df9 = preprocess_dataset(dataset9, ['questionTitle', 'questionText','answerText','text' ])

print("Datasets ready to be pre-processed")
```

## **Data Pre-processing**

```python
# Define a balanced preprocessing function
import re

def balanced_preprocessing(text):
    # Lowercase the text
    text = text.lower()
    # Remove URLs
    text = re.sub(r'http\S+', '', text)
    # Replace non-alphanumeric characters with spaces
    text = re.sub(r'[^a-zA-Z0-9\s]', ' ', text)
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

```python
# Apply the preprocessing function to the text column
df1['preprocessed_text'] = df1['text'].apply(balanced_preprocessing)
df2['preprocessed_text'] = df2['text'].apply(balanced_preprocessing)
df3['preprocessed_text'] = df3['text'].apply(balanced_preprocessing)
df4['preprocessed_text'] = df4['text'].apply(balanced_preprocessing)
df5['preprocessed_text'] = df5['text'].apply(balanced_preprocessing)
df6['preprocessed_text'] = df6['text'].apply(balanced_preprocessing)
df7['preprocessed_text'] = df7['text'].apply(balanced_preprocessing)
df8['preprocessed_text'] = df8['text'].apply(balanced_preprocessing)
df9['preprocessed_text'] = df9['text'].apply(balanced_preprocessing)
print("Datasets are ready to get combined!")
```

```python
# Display the first few rows of the original and preprocessed text
print("Original vs Preprocessed Text:")
print(df1[['text', 'preprocessed_text']].head())
```

```python
# Concatenate all preprocessed DataFrames
combined_df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9], ignore_index=True)
combined_df.shape
```
